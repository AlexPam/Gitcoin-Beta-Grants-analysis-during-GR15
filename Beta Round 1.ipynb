# Using the Website Validation Lego.

```python
# Importing relevant libraries
import os
import pandas as pd
import requests
import math
```

```python
#A Function to to check the status of a website. If the function returns a status code of 200 it means it is valid
#Else, if it returns any other status code it is invalid.
#I also modified the website lego to return no connetion for websites with no response

def website_validation(website):
    website_working = []

    try:
        website_status = requests.get(website, timeout=10, verify=False).status_code
        if website_status == 200:
            website_working.append(True)
        else:
            website_working.append(False)
    except requests.exceptions.RequestException:
        website_working.append('no_connection')

    return website_working[0]
```

```python
# Renaming the metadata.website column as website to avoid error because of the '.' character
first_ten_cols = first_ten_cols.rename(columns={'metadata.website':'website','metadata.title':'title','metadata.userGithub':'userGithub','metadata.projectGithub':'projectGithub','metadata.description':'projectDescription'})
```

```python
#Data Preview
first_ten_cols.head(10)
```

```python
# Running a for loop to check the status of all grantee websites
result = []
for i in first_ten_cols['website']:
    result.append(website_validation(i))
```

```python
# Saving the output as 'result'
first_ten_cols['website_validation_lego'] = result
```

# Using the Github Lego

```python
"""" this code defines a function that uses the Github API to validate a Github account and returns True if the account is valid, and False otherwise."""

def github_username_validation(userGithub):
    github_api_url = f"https://api.github.com/users/{userGithub}"
    try:
        response = requests.get(github_api_url, timeout=10)
        if response.status_code == 200:
            return True
        elif response.status_code == 404:
            return False
        else:
            return False
    except requests.exceptions.RequestException:
        return False
```

```python
# Running a for loop to check the status of all grantee username
result = []
for i in first_ten_cols['userGithub']:
    result.append(github_username_validation(i))
```

```python
# Saving the output as 'result'
first_ten_cols['github_account_validation'] = result
```
